```python
import pandas as pd
```

```python
cwc_df = pd.read_csv('CWC ATTENDANCE LIST main.csv')

cwc_df
```

```python
nan_df = cwc_df[cwc_df.isna().any(axis=1)] #Showing members who do not have a complete profile on the register
nan_df.head(60)
```

```python
cwc_df = cwc_df[cwc_df['LEVEL'].notna()]

cwc_df.head(100)
```

```python
cwc_df.loc[cwc_df['E-MAIL ADDRESS'] == 'Sowauzamere@gmail.com', ['DATE OF BIRTH']] = 'Februrary 29th'

#Realised Sowa's date of birth was blank so I decided to fill it manually.
```

```python
cwc_df
```

```python
groups = cwc_df.groupby('LEVEL')

groups.count() #Counts the number of people belonging to that level

df_count = groups.count()

df_count
```

```python
import matplotlib.pyplot as plt
        
levels = [level for level, df in cwc_df.groupby('LEVEL')] #In order for the city to be arranged in order with thier respective sales

plt.bar(levels, df_count['NAME'], width = 50)

plt.xticks(levels) 
plt.ylabel('Number of members')
plt.xlabel('Level')
plt.show()
```

```python
cwc_df ['Month'] = cwc_df['DATE OF BIRTH'].str[0:3]
cwc_df.head()
```

```python
bdays = cwc_df.groupby('Month')

bdays.count() #Counts the number of people belonging to that level

bday_count = bdays.count()

bday_count
```

```python
import matplotlib.pyplot as plt
        
months = [month for month, df in bdays] #In order for the city to be arranged in order with thier respective sales

plt.bar(months, bday_count['NAME'])
plt.plot(months, bday_count['NAME'], color = 'g')

plt.xticks(months) 
plt.ylabel('Number of Members')
plt.xlabel('Month of Birth')
plt.grid()
plt.show()
```
